#### Correlation analysis

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler

iris_copy = iris.copy()
## converting species to numericals
led = LabelEncoder()
iris_copy['species'] = led.fit_transform(iris_copy['species'])
```

```python
iris_copy.corr()
```

```python
iris_copy['species'].value_counts()
```

```python
import seaborn as sns

sns.heatmap(iris_copy.corr(), annot = True, center = 0, cmap = 'Blues')
```

#### Model selection Training and Testing

```python
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier


X = iris_copy.drop('species', axis = 1)
y = iris_copy['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)

knn = KNeighborsClassifier()

scores_ = cross_val_score(knn, X_train, y_train, cv = 5)

print("Cross validation scores: ", scores_)

print("Cross validation mean score: ", scores_.mean())
```

```python
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
y_pred
```

```python
test_score = knn.score(X_test, y_test)
test_score
```

```python
knn.predict([[8.0, 3.5, 3.0, 1.9]])
```

```python
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay, confusion_matrix

acc_score = accuracy_score(y_test, y_pred)
acc_score
```

```python
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
p = ConfusionMatrixDisplay(confusion_matrix = cm)
p.plot()
```

```python
param_grid = {'n_neighbors': range(1, 20)}

grid_search = GridSearchCV(knn, param_grid, cv=5)

grid_search.fit(X_train, y_train)


print(grid_search.best_params_)

print(grid_search.best_score_)


best_knn = grid_search.best_estimator_

best_knn.fit(X_train, y_train)

y_new_pred = best_knn.predict(X_test)
print()
print("accuracy score: ", accuracy_score(y_test, y_new_pred))
```

```python
print(classification_report(y_test, y_new_pred))
```
